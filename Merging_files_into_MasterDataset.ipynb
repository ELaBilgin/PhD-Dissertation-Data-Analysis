#Read files

```python
import pandas as pd

prs = pd.read_excel("PRS_PrimaryOutcomes.xlsx")
eeg = pd.read_excel("EEG_PrimaryOutcomes.xlsx")
sart = pd.read_excel("SART2_PrimaryOutcomes.xlsx")

print("PRS data:")
display(prs.head())

print("EEG data:")
display(eeg.head())

print("SART2 data:")
display(sart.head())
```

#Check ParticipantID column

```python
for name, data in {
    "PRS": prs,
    "EEG": eeg,
    "SART2": sart
}.items():

    if "ParticipantID" not in data.columns:
        print(f"ParticipantID column is missing in {name} file.")
    else:
        print(f"ParticipantID column found in {name} file.")
```

# Merge all datasets by ParticipantID

```python
master = prs.merge(
    eeg,
    on="ParticipantID",
    how="inner"
)

master = master.merge(
    sart,
    on="ParticipantID",
    how="inner"
)

master = master.sort_values("ParticipantID")

print("Master dataset created successfully.")
display(master)
```

# Download master dataset

```python
master_file = "MasterDataset.xlsx"

master.to_excel(
    master_file,
    index=False
)

print(f"{master_file} saved.")
```
